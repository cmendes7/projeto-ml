```python
import pandas as pd
df = pd.read_csv("preco_imovel.csv")
df
```

```python
df.plot(kind="scatter", x="metros_quadrados", y="preco");
```

```python
from sklearn import linear_model
modelo = linear_model.LinearRegression()
x = df[["metros_quadrados"]]
y = df[["preco"]]

modelo.fit(x, y)
```

```python
print(modelo.predict([[200]])[0][0])
```
