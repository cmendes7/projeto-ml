```python
import pandas as pd
df = pd.read_csv("hectare_terreno.csv")
df
```

```python
df.plot(kind="scatter", x="hectare", y="preco");
```

```python
from sklearn import linear_model
modelo = linear_model.LinearRegression()
x = df[["hectare"]]
y = df[["preco"]]

modelo.fit(x, y)
```
